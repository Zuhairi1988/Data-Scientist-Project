## FEATURE ENGINEERING

```python
x = df[['number of bathrooms','grade of the house', 'living area', 'Area of the house(excluding basement)', 'living_area_renov', 'number of views', 'Area of the basement', 'number of bedrooms', 'Lattitude', 'Longitude','Built Year', 'Renovation Year', 'waterfront present', 'condition of the house', 'lot_area_renov']]
y = df[['Price']]
```

### DATA TRAINING

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
df.shape[0] 
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["mse", "friedman_mse", "mae"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" :[2,5,10],
    "min_samples_leaf" : [1,2,4]
}
```

```python
tree_model =DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, tree_preds)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predlr = lr.predict(x_test)
```

```python
mean_squared_error(y_test, predlr)
```

```python
mean_absolute_error(y_test, predlr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfmodel, param_gridrfr)
```

```python
gridrfr.fit(x_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(x_test)
```

```python
mean_absolute_error(y_test,rfrpredictions)
```
